```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

df = pd.read_csv('/Users/zhehaoyu/Desktop/MCM/D题/2020_Problem_D_DATA/passingevents.csv') # 传球
df = df.sort_values(['MatchID', 'OriginPlayerID', 'DestinationPlayerID'])
df
```

```python
# 计算哈士奇的gamewise k
passes = df[['MatchID','TeamID', 'OriginPlayerID', 'EventOrigin_x', 'EventOrigin_y', 'EventDestination_x', 'EventDestination_y']]
passes['huskies_dx'] = passes['EventDestination_x'] - passes['EventOrigin_x']
passes['huskies_dy'] = abs(passes['EventDestination_y'] - passes['EventOrigin_y'])
```

```python
passes31 = passes[passes['MatchID']==31]
```

```python
passes31 = passes31[passes31['TeamID']=='Huskies']
k = passes31.groupby('OriginPlayerID').mean()
k['k'] = k['huskies_dy'] / k['huskies_dx']
k['1/k'] = 1/k['k']
k.to_csv('k_in_game_31.csv',index=True,sep=',')
```

```python
k.describe()
```

```python
Hpass = passes[passes['TeamID']=='Huskies']
Hpass = Hpass.groupby(['MatchID']).mean()
Hpass = Hpass[['huskies_dx','huskies_dy']]
Hpass['huskies_k'] = Hpass['huskies_dy'] / Hpass['huskies_dx']
```

```python
Hpass.to_csv('hskies_k_in_each_game.csv',index=True,sep=',')
```

```python
# 计算全局的球员传球位置
subdf = df[['OriginPlayerID', 'EventOrigin_x', 'EventOrigin_y']]
location = subdf.groupby(['OriginPlayerID']).mean()
location.to_csv('macro_location.csv',index=True,sep=',')
```

```python
# 计算k
passes = df[['MatchID','TeamID', 'OriginPlayerID', 'EventOrigin_x', 'EventOrigin_y', 'EventDestination_x', 'EventDestination_y']]
passes['dx'] = passes['EventDestination_x'] - passes['EventOrigin_x']
passes['dy'] = abs(passes['EventDestination_y'] - passes['EventOrigin_y'])
direction = passes.groupby(['MatchID','TeamID']).mean()
direction
dire = direction[['dx','dy']]
dire['k'] = dire['dy'] / dire['dx']
dire
```

```python
dire.to_csv('opponents_k.csv',index=True,sep=',')
```
