```python
df = pd.read_csv('/Users/zhehaoyu/Desktop/MCM/D题/2020_Problem_D_DATA/passingevents.csv') # 传球
df = df.sort_values(['MatchID', 'OriginPlayerID', 'DestinationPlayerID'])
df
```

```python
#计算每个边权值
cnt = df.groupby(['MatchID', 'OriginPlayerID', 'DestinationPlayerID']).count()
cnt = cnt['TeamID']
cnt
```

```python
import math

event = pd.read_csv('/Users/zhehaoyu/Desktop/MCM/D题/2020_Problem_D_DATA/fullevents.csv')
ev = event.sort_values(['MatchID','EventSubType','TeamID'])
ev = event[(event['EventType']=='Pass')& (event['TeamID']=='Huskies')] # & math.isnan(event['DestinationPlayerID'])]
ev = ev.sort_values(['MatchID','OriginPlayerID'])
# fault = ev[ev['DestinationPlayerID'].isnull()]
# fgroup = fault.groupby(['MatchID','OriginPlayerID']).count()
# fnumber = fgroup['TeamID']
# fnumber
pnumber = ev.groupby(['MatchID','OriginPlayerID']).count()
pass_data = pnumber[['TeamID','DestinationPlayerID']]
pass_data.columns = ['total','success']
pass_data['rate'] = pass_data['success']/pass_data['total']
pass_data.to_csv('pass_success_rate.csv',index=True,sep=',')
```
