```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
import math

event = pd.read_csv('/Users/zhehaoyu/Desktop/MCM/D题/2020_Problem_D_DATA/fullevents.csv')
ev = event.sort_values(['MatchID','EventSubType','TeamID'])
ev = event[(event['EventType']=='Pass')& (event['TeamID']=='Huskies')] # & math.isnan(event['DestinationPlayerID'])]
ev = ev.sort_values(['MatchID','OriginPlayerID'])
# fault = ev[ev['DestinationPlayerID'].isnull()]
# fgroup = fault.groupby(['MatchID','OriginPlayerID']).count()
# fnumber = fgroup['TeamID']
# fnumber
pnumber = ev.groupby(['MatchID','OriginPlayerID']).count()
Hpass_data = pnumber[['TeamID','DestinationPlayerID']]
Hpass_data.columns = ['total','success']
Hpass_data['rate'] = Hpass_data['success']/Hpass_data['total']
Hpass_data_in_each_game = Hpass_data.groupby(['MatchID']).sum()
Hpass_data_in_each_game['rate'] = Hpass_data_in_each_game['success']/Hpass_data_in_each_game['total']
```

```python
Hpass_data_in_each_game.describe()
```

```python
ev2 = event[(event['EventType']=='Pass')& (event['TeamID']!='Huskies')] # & math.isnan(event['DestinationPlayerID'])]
ev2 = ev2.sort_values(['MatchID','OriginPlayerID'])
# fault = ev[ev['DestinationPlayerID'].isnull()]
# fgroup = fault.groupby(['MatchID','OriginPlayerID']).count()
# fnumber = fgroup['TeamID']
# fnumber
Opponent_pnumber = ev2.groupby(['MatchID','OriginPlayerID']).count()
Opass_data = Opponent_pnumber[['TeamID','DestinationPlayerID']]
Opass_data.columns = ['total','success']
Opass_data['rate'] = Opass_data['success']/Opass_data['total']
Opass_data_in_each_game = Opass_data.groupby(['MatchID']).sum()
Opass_data_in_each_game['rate'] = Opass_data_in_each_game['success']/Opass_data_in_each_game['total']
```

```python
Opass_data_in_each_game = Opass_data_in_each_game.rename(columns={'total':'Ototal', 'success':'Osuccess', 'rate':'Orate'})
Opass_data_in_each_game
```

```python
pass_data_in_each_game=pd.concat([Hpass_data_in_each_game,Opass_data_in_each_game],axis=1)
pass_data_in_each_game.to_csv('pass_data_in_each_game.csv',index=True,sep=',')
```
