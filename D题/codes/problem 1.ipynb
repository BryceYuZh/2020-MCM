```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
df = pd.read_csv('/Users/zhehaoyu/Desktop/MCM/Dé¢˜/2020_Problem_D_DATA/passingevents.csv')
```

```python
match1 = df[(df['MatchID']==1)&(df['TeamID']=='Huskies')]
match2 = df[(df['MatchID']==2)&(df['TeamID']=='Huskies')]
match3 = df[(df['MatchID']==3)&(df['TeamID']=='Huskies')]
```

```python
pass1 = match1[['EventOrigin_x','EventOrigin_y','EventDestination_x','EventDestination_y']]
pass2 = match2[['EventOrigin_x','EventOrigin_y','EventDestination_x','EventDestination_y']]
pass3 = match3[['EventOrigin_x','EventOrigin_y','EventDestination_x','EventDestination_y']]
```

```python
zoom =[]
for i,row in pass1.iterrows():
    if row['EventOrigin_x'] >= 0 and row['EventOrigin_x'] <= 20:
        zoom.append(1)
    if row['EventOrigin_x'] > 20 and row['EventOrigin_x'] <= 40:
        zoom.append(2)
    if row['EventOrigin_x'] > 40 and row['EventOrigin_x'] <= 60:
        zoom.append(3)
    if row['EventOrigin_x'] > 60 and row['EventOrigin_x'] <= 80:
        zoom.append(4)
    if row['EventOrigin_x'] > 80 and row['EventOrigin_x'] <= 100:
        zoom.append(5)

pass1['room'] = zoom
pass1['huskies_dx'] = pass1['EventDestination_x'] - pass1['EventOrigin_x']
pass1['huskies_dy'] = abs(pass1['EventDestination_y'] - pass1['EventOrigin_y'])
pass1['huskies_k'] = pass1['huskies_dy'] / pass1['huskies_dx']
result = pass1[['room','huskies_k']]
result
result.to_csv('short_time_k vs. room.csv',index=True,sep=',')
```
