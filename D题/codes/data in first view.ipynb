```python
grouped = df.groupby(['MatchID','TeamID']).count()
grouped
pass_info = grouped['OriginPlayerID']
pass_info
match = np.arange(1,39)
Hpass = []
Opass = []

cnt = 1
for i, v in pass_info.iteritems():
    if cnt%2==1:
        Hpass.append(v)
    else:
        Opass.append(v)
    cnt = cnt+1
```
