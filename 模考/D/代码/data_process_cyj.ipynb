```python
from geopy.distance import geodesic
```

```python
import numpy
import pandas
df = pandas.read_csv('/Users/zhehaoyu/Desktop/MCM/模考/D/代码/1/order_20161101',encoding="utf-8")
df = df.drop(['eb9dd4095d9850e6287cefd813775a6c'],axis=1)
df.columns = ['times','timee','sl','ss','el','es']
df.describe()
```

```python
data = df.loc[:,['sl','ss']]
data2= df.loc[:, ['el','es']]
data2.rename(columns={'el':'sl', 'es':'ss'}, inplace = True)
data = data2.append(data, ignore_index = False)
data.describe()
```

```python
def dist (lon1, lat1, lon2, lat2):
    return geodesic((lat1,lon1),(lat2,lon2)).km
```

```python
class Hotpoint:
    def __init__(self, lon, lat):
        self.lon = lon
        self.lat = lat
        self.around = 0

class Endpoint:
    def __init__(self, lon, lat):
        self.lon = lon
        self.lat = lat
        self.around = 0
```

```python
# calculate end point
ed = []
for index, row in df.iterrows():
    flag = True 
    for v in ed:
        if dist(row['el'], row['es'], v.lon, v.lat) < 1 :
            flag = False
            break
    if flag:
        ed.append(Endpoint(row['el'],row['es']))
```

```python
vs = []
for index, row in data.iterrows():
    flag = True 
    for v in vs:
        if dist(row['sl'], row['ss'], v.lon, v.lat) < 2 :
            flag = False
            break
    if flag:
        vs.append(Hotpoint(row['sl'],row['ss']))
```

```python
dt = data[1:848143:50]
```

```python
# calculate probability
for v in vs:
    cnt = 0
    for index, row in dt.iterrows():
        if dist(row['sl'], row['ss'], v.lon, v.lat) < 2 :
            cnt = cnt+1
        v.around = cnt
```

```python
for v in ed:
    cnt = 0
    for index, row in dt.iterrows():
        if dist(row['el'], row['es'], v.lon, v.lat) < 1:
            cnt = cnt+1
        v.around = cnt
```
